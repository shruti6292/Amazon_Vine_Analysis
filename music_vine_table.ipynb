```python
# import dependancies
import pandas as pd
```

```python
# reviews file to load
music_vine_df = pd.read_csv("music_vine_table.csv")
music_vine_df
```

```python
# explore the dataframe
music_vine_df.info()
```

## Step 1 
Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20 to pick reviews that are more likely to be helpful and to avoid having division by zero errors later on.

```python
# filtering all rows where total_votes >=20
music_vine_20_df = music_vine_df[music_vine_df["total_votes"] >= 20]
music_vine_20_df
```

## Step 2
Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.

```python
# all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
music_vine_50_df = music_vine_20_df.loc[(music_vine_20_df["helpful_votes"] / music_vine_20_df["total_votes"] >= 0.5)]
music_vine_50_df
```

# Step 3
Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.

```python
# filter vine_50_df where vine= Y : paid reviews
music_vine_paid_reviews_df = music_vine_50_df.loc[music_vine_50_df["vine"] == 'Y']
music_vine_paid_reviews_df
```

# Step 4
Repeat Step 3, but this time retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.

```python
# filter vine_50_df where vine= N : unpaid reviews
music_vine_unpaid_reviews_df = music_vine_50_df.loc[music_vine_50_df["vine"] == 'N']
music_vine_unpaid_reviews_df
```

# Step 5
Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid)

```python
# determine the count of paid 5 star reviews
paid_five_star_reviews = music_vine_paid_reviews_df.loc[music_vine_paid_reviews_df["star_rating"] == 5]["star_rating"].count()
paid_five_star_reviews
```

```python
# determine the count of paid reviews
paid_reviews = music_vine_paid_reviews_df["star_rating"].count()
paid_reviews
```

```python
# determine the % of paid 5 * reviews
percentage_paid_five_star_reviews = (paid_five_star_reviews / paid_reviews) * 100
percentage_paid_five_star_reviews
```

```python
# determine the count of unpaid 5 star reviews
unpaid_five_star_reviews = music_vine_unpaid_reviews_df.loc[music_vine_unpaid_reviews_df["star_rating"] == 5]["star_rating"].count()
unpaid_five_star_reviews
```

```python
# determine the count of unpaid reviews
unpaid_reviews = music_vine_unpaid_reviews_df["star_rating"].count()
unpaid_reviews
```

```python
# determine the % of paid 5 * reviews
percentage_unpaid_five_star_reviews = (unpaid_five_star_reviews / unpaid_reviews) * 100
percentage_unpaid_five_star_reviews
```
